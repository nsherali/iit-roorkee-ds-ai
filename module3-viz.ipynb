```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("/content/cars (1) (1).csv")
df
```

```python
df.info()
```

1. Create scatter plots for the following data, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.

a. Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg, hp-time-to-60) with respect to each brand.

```python
fig = plt.subplots(2,2,figsize=(30,20))


plt.subplot(2,2,1)     #subplot(row,col,position)
sns.scatterplot(x=df['mpg'],y=df['hp'],hue=df['brand'])
plt.title('MPG VS HP')
plt.xlabel('MPG')
plt.ylabel('HP')


plt.subplot(2,2,2)
sns.scatterplot(x=df['mpg'],y=df['weightlbs'],hue=df['brand'])
plt.title('MPG VS weightlbs')
plt.xlabel('MPG')
plt.ylabel('weightlbs')


plt.subplot(2,2,3)
sns.scatterplot(x=df['hp'],y=df['mpg'],hue=df['brand'])
plt.title('MPG VS HP')
plt.ylabel('MPG')
plt.xlabel('HP')

plt.subplot(2,2,4)
sns.scatterplot(x=df['hp'],y=df['time-to-60'],hue=df['brand'])
plt.title('time-to-60 VS HP')
plt.ylabel('time-to-60')
plt.xlabel('HP')


plt.show()
```

Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands.

```python
fig = plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.barplot(y=df['hp'],x=df['cylinders'],hue=df['brand'])

plt.subplot(2,2,2)
sns.barplot(y=df['mpg'],x=df['cylinders'],hue=df['brand'])

plt.subplot(2,2,3)
sns.barplot(y=df['weightlbs'],x=df['cylinders'],hue=df['brand'])

plt.subplot(2,2,4)
sns.barplot(y=df['time-to-60'],x=df['cylinders'],hue=df['brand'])

plt.show()
```

Create box plots for the columns in the step1 (hp, mpg, weight, time-to-60) with respect to the number of cylinders and brand respectively.

```python
fig = plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.boxplot(y=df['hp'],x=df['cylinders'],hue=df['brand'])

plt.subplot(2,2,2)
sns.boxplot(y=df['mpg'],x=df['cylinders'],hue=df['brand'])

plt.subplot(2,2,3)
sns.boxplot(y=df['weightlbs'],x=df['cylinders'],hue=df['brand'])

plt.subplot(2,2,4)
sns.boxplot(y=df['time-to-60'],x=df['cylinders'],hue=df['brand'])

plt.show()
```

a. Create histograms for hp, mpg, time-to-60, weight lbs, cubic inches, year, with respect to brand.

```python
fig = plt.subplots(3,2,figsize=(20,10))

plt.subplot(3,2,1)
sns.histplot(x=df['hp'],hue=df['brand'])

plt.subplot(3,2,2)
sns.histplot(x=df['mpg'],hue=df['brand'])

plt.subplot(3,2,3)
sns.histplot(x=df['time-to-60'],hue=df['brand'])

plt.subplot(3,2,4)
sns.histplot(x=df['weightlbs'],hue=df['brand'])

plt.subplot(3,2,5)
sns.histplot(x=df['cubicinches'],hue=df['brand'])

plt.subplot(3,2,6)
sns.histplot(x=df['year'],hue=df['brand'])

plt.show()
```

 Line plot between hp-mpg, hp-(time-to-60), mpg-(time-to-60), hp-weight lbs, mpg-weight lbs, cubic inches-weight lbs.

```python
fig = plt.subplots(2,3,figsize=(20,10))

plt.subplot(2,3,1)
sns.lineplot(x=df['hp'],y=df['mpg'],hue=df['brand'])

plt.subplot(2,3,2)
sns.lineplot(x=df['hp'],y=df['time-to-60'],hue=df['brand'])

plt.subplot(2,3,3)
sns.lineplot(x=df['mpg'],y=df['time-to-60'],hue=df['brand'])

plt.subplot(2,3,4)
sns.lineplot(x=df['hp'],y=df['weightlbs'],hue=df['brand'])

plt.subplot(2,3,5)
sns.lineplot(x=df['cubicinches'],y=df['weightlbs'],hue=df['brand'])

plt.subplot(2,3,6)
sns.lineplot(x=df['mpg'],y=df['weightlbs'],hue=df['brand'])

plt.show()
```

```python
colx=['hp','hp','mpg','hp','cubicinches','mpg']
coly=['mpg','time-to-60','time-to-60','weightlbs','weightlbs','weightlbs']

fig=plt.subplots(2,3,figsize=(20,10))

for i in range(len(colx)):
  plt.subplot(2,3,i+1)
  sns.lineplot(x=df[colx[i]],y=df[coly[i]],hue=df['brand'])
plt.show()
```
